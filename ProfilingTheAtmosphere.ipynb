## Stratification of the Atmosphere

The stratification of atmosphere is contiously monitored all over the planet in the frame of internationally coordinated athmospheric sounding programms. There exisit a large number of methods to measure improtant state variables such as temperature, humidtiy, wind speed, trace gas concentrations (further reading: https://www.sciencedirect.com/topics/earth-and-planetary-sciences/atmospheric-sounding).

Actual athmospheric data are available through different online platformes (e.g. http://weather.uwyo.edu/upperair/sounding.html).

In the following we will read in data from radiosondes lauched close to Berlin (Lichtenberg) and plot the records.

```python
#read HDF5 file which contains atmopheric sounding data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

hdf = pd.read_hdf('data/profile_day.h5')
```

```python
#show dataframe
hdf
```

| Parameter  |              Description               |      Units         |
|:----------:|:--------------------------------------:|:------------------:|
|   PRES:    | Atmospheric Pressure                   | [hPa]              |
|   HGHT:    | Geopotential Height                    | [meter]            |
|   TEMP:    | Temperature                            | [celsius]          |
|   DWPT:    | Dewpoint Temperature                   | [celsius]          |
|   FRPT:    | Frost Point Temperature                | [celsius]          |
|   RELH:    | Relative Humidity                      | [%]                |
|   RELI:    | Relative Humidity with respect to Ice  | [%]                |
|   MIXR:    | Mixing Ratio                           | [gram/kilogram]    |
|   DRCT:    | Wind Direction                         | [degrees true]     |
|   SKNT:    | Wind Speed                             | [knot]             |
|   THTA:    | Potential Temperature                  | [kelvin]           |
|   THTE:    | Equivalent Potential Temperature       | [kelvin]           |
|   THTV:    | Virtual Potential Temperature          | [kelvin]           |

```python
#plot air temperature and dew point temperature as red and blue lines

fig, ax1 = plt.subplots(figsize=(10,6))
ax1.plot(hdf['TEMP'], hdf['HGHT'], color='red')
ax1.set_ylabel('Geopotential Height [m]')
ax1.set_xlabel('Temperature')
ax1.set_ylim(0,3000)
#ax1.set_xlim(290,300)

ax2 = ax1.twiny()
ax2.plot(hdf['MIXR'], hdf['HGHT'], color='blue')
ax2.set_xlabel('Releative Humidity')

fig.tight_layout()
plt.show()
```
